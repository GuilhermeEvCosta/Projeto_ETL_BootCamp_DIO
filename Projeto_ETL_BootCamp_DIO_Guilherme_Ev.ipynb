# Projeto ETL | BootCamp DIO
# Campanhas especificas com perfil do cliente.

**Contexto:** Com base no saldo geral dos clientes (Conta Corrente/Poupança), é instituido em qual campanha o mesmo se enquadra, e qual comunicação receberá, sendo elas:

1.   Consórcio é aqui! | Antecipe seu sonho com nosso consórcio!
2.   Antecipa FGTS!	   | Está no sufoco? Antecipe seu FGTS conosco!
3.   Invista Conosco!  | Invista conosco e faça seu dinheiro render mais!

**Etapas:**
Extrair:
1. Extrair base csv com os dados base para o exercicio.

Transformar:
2. Identificar o grupo de acordo com saldo.
3. Definir mensagem.

Carregar:
4. Uma vez que a mensagem para cada cliente esteja pronta, salvar as informações em formato csv, e disponibilizando o material enriquecido.

# E | Extrair base csv.

```python
# Instalação Bibliotecas
import pandas as pd
import numpy as np

# Definindo Data Frame
df = pd.read_csv('banco_users.csv')
```

```python
# Apresentação Dta Frame
df
```

# T | Transformar dados.

```python
# Criando coluna 'Saldo_total', da soma de 'Saldo_Poupança' e 'Saldo_corrente'
df['Saldo_total'] = df['Saldo_corrente'] + df['Saldo_poupanca']

# Aprensentação
print(df)
```

```python
# Função para atribuir grupos com base no valor da coluna 'Saldo_total'

def atribuir_campanha(saldo_total):
    if saldo_total <= 1.0:
        return 'Antecipa FGTS!'
    elif 1.001 <= saldo_total <= 10.000:
        return 'Consórcio é aqui!'
    elif 10.001 <= saldo_total <= 100.000:
        return 'Invista Conosco!'

# Aplicar a função à coluna 'Saldo_total' para criar a coluna 'Campanha_User'
df['Campanha_User'] = df['Saldo_total'].apply(atribuir_campanha)

# Apresentação Df
print(df)
```

```python
# Apresentação Df
df.head()
```

```python
# Função para gerar mensagens de marketing com base na coluna 'Campanha_User'
def mensagem_de_marketing(campanha):
    if campanha == 'Antecipa FGTS':
        return 'Está no sufoco? Antecipe seu FGTS conosco!'
    elif campanha == 'Consórcio é aqui!':
        return 'Antecipe seu sonho com nosso consórcio!'
    elif campanha == 'Invista Conosco!':
        return 'Invista conosco e faça seu dinheiro render mais!'

# Aplicar a função à coluna 'Campanha_User' para criar a coluna 'Mensagem_Marketing'
df['Mensagem_Marketing'] = df['Campanha_User'].apply(mensagem_de_marketing)

# Limitar as mensagens a 100 caracteres
df['Mensagem_Marketing'] = df['Mensagem_Marketing'].str[:100]

# Apresentação Df com as informações da Campanha e Mensagem a ser enviada
print(df[['Campanha_User', 'Mensagem_Marketing']])
```

```python
# Função para mapear as chaves de mensagem com base na coluna 'Campanha_User'
def mapear_chave_mensagem(campanha):
    if campanha == 'Antecipa FGTS':
        return 'msg1'
    elif campanha == 'Consórcio é aqui!':
        return 'msg2'
    elif campanha == 'Invista Conosco!':
        return 'msg3'

# Aplicar a função à coluna 'Campanha_User' para criar a coluna 'Chave_Mensagem'
df['Chave_Mensagem'] = df['Campanha_User'].apply(mapear_chave_mensagem)

# Apresentação das colunas Campanha e Chave
print(df[['Campanha_User', 'Chave_Mensagem']])
```
